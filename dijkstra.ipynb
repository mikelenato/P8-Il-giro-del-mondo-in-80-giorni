# Calcola lo shortest path con l'algoritmo di Dijkstra su un grafo orientato weighted calcolato con networkx

```python
import numpy as np 
import pandas as pd
pd.set_option('display.max_colwidth', 4000)
path = 'C:\\Users\michele.lenato\Downloads\\input.xlsx'  # Inserire il path in cui è stato scaricato il file input.xlsx del repo github
input = pd.read_excel(path) 
del input['Unnamed: 0']
source = 1826657058  # Holborn
target = 1826645935  # Londra
nodes_AZ = input['A'].tolist() + input['Z'].tolist()
nodes = set(nodes_AZ) #26567 nodi del grafo
```

```python
s = set(nodes)-set([target, source])
n = [source] + list(s)
n.append(target) 
```

Calcolo grafo con networkx

```python
#calcolo grafo direzionato
input_for_dg = []
for i in range(input.shape[0]):
  input_for_dg.append(tuple(input.loc[i].values.tolist()))
import networkx as nx
DG = nx.DiGraph()
DG.add_weighted_edges_from(input_for_dg)
```

Matrice delle adiacenze

```python
from datetime import datetime
now = datetime.now()
current_time = now.strftime("%H:%M:%S")
print("Current Time =", current_time)

ADJ = np.zeros((len(n),len(n)))
for i in range(len(n)):
  for j in range(len(n)):
    if n[i] == n[j]:
      ADJ[i,j] = 0  # se A-Z coincidono
    elif n[i] != n[j]:
      try:
        ADJ[i,j] = ((DG[n[i]][n[j]]['weight']))
      except KeyError:
        ADJ[i,j] == 99999999999999  # se A-Z non sono connessi 
```

Inizializzazione

```python
previous = list(range(len(n))) # i nodi vengono mappati su indici da 0 a 26566
d = np.array([0]+[None]*len(n)) # vettore delle distanze tra source e gli altri nodi, iniziale
pred = np.array(previous) # è la lista dei nodi "best" che precedono l'ultimo nodo analizzato, all'inizio coincidono con se stessi
grafo = ADJ #matrice adiacenze
```

Iterazione. Output: la lista "pred" con i predecessori "best" a ritroso dal target al source

```python
from datetime import datetime
now = datetime.now()
current_time = now.strftime("%H:%M:%S")
print("Current Time =", current_time)

L = set([0]) #lista dei nodi A da analizzare, inizialmente c'è il nodo source
S = set([]) #lista nodi analizzati, inizialmente vuoto 
while (len(L) > 0):
  z = None 
  x = None 
  for i in L:
    if ((z == None) or (d[i] < z)):
      z = d[i]  # il peso del nodo iniziale coincide col peso dell'edge dal nodo più vicino
      x = i    #  # seleziona un nodo x da L con distanza minore dal nodo iniziale 
  L.remove(x) 
  S.add(x) 
  k = 0
  for i in grafo[x]: # scansione dei nodi adiacenti al nodo tolto da L
    if ((i > 0) & (k not in S)): 
      node_ad = k 
      L.add(node_ad) 
      if ((d[k] == None) or (d[x] + grafo[x,k] < d[k])):
        d[k] = d[x] + grafo[x,k] 
        pred[k] = x
    k = k + 1
print("nodo=",x,", L=",L,", S=",S,", d=", d,", pred=", pred)
```

Calcolo shortest path. Input: la precedente lista "pred"

```python
#sp: nodi=pred[ultimo indice] in poi. 
lista_sp = [26566] #la lista all'inizio contiene l'indice del nodo target, Londra. Completa la lista_sp alla fine della ricorsione
def shortest_path(last):
  global lista_sp
  if pred[last] == 0:
    lista_sp.append(0)
    return lista_sp
  prossimo = pred[last]
  lista_sp.append(prossimo)
  return shortest_path(prossimo)
```

```python
# dagli indici ricavo i nodi del grafo
nodi = []
for i in lista_sp:
  nodi.append(n[i])
```

```python
#shrotest path con dijkstra
sh_path = list(reversed(nodi))
sp_dij = pd.DataFrame(sh_path, columns=['shortest_path_dijkstra'])
sp_dij.to_excel('C:\\Users\michele.lenato\Downloads\shortest_path_dijkstra.xlsx') # da importare nel main()
```
